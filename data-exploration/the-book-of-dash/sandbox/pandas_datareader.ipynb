```python
import pandas_datareader.data as web
import datetime
```

```python
start = datetime.datetime(2022, 1, 1)
end = datetime.datetime.today()
close = web.DataReader(['SIE.DE'], 'stooq', start, end).Close
close.plot()
```

```python
start = datetime.datetime(2022, 1, 1)
end = datetime.datetime(2022, 12, 3)
df = web.DataReader(['AMZN','GOOGL','GE','PFE','MRNA','BNTX'],
                    'stooq', start=start, end=end).Close

df.plot(subplots=True, layout=(2, 3))
```
