# Chapter 1 Time Series Concepts and Python

```python
import pandas as pd
import numpy as np
import datetime as dt
import xlrd # convert excel date series to datetime
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style()
```

```python
date_parser = lambda x: xlrd.xldate_as_datetime(x, 0)
df = pd.read_excel(
    "data/BitcoinPrice17-6-23-18-6-22.xlsx", index_col=0, usecols=[0, 1],
    parse_dates=True, date_parser=date_parser
)
df.plot();
```

```python
val = np.ravel(pd.read_excel('data/milk.xlsx', header=None).values)
val = val[~np.isnan(val)]
milk = pd.Series(index=pd.date_range('19620101', periods=len(val), freq='M'), data=val)
milk.plot();
```

```python
birth = pd.read_csv('data/Noboyngirl.csv', index_col=1)
fig, ax = plt.subplots()
ax.plot(birth.boys, label='boys')
ax.plot(birth.girls, label='girls')
fig.legend(loc=(0.8, 0.1))
plt.show()
```
